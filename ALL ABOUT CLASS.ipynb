```python
class Car:
    #constructor
    def __init__(self,brand,color,price):
        self.brand = brand
        self.color = color
        self.price = price
    def start(self):
        print("Car started....")

    def car_brand(self):
        print("Car Brand",self.brand)
    def car_color(self):
        print("Car color",self.color)
    def car_price(self):
        print("Car price",self.price)

my_car = Car("Honda","Gray",1200000)
my_car.start()
my_car.car_brand()
my_car.car_color()
my_car.car_price()
```
